find average marks

```python
students = [
    {"id": 1, "name": "sreerekha", "marks": [85, 90, 88]},
    {"id": 2, "name": "athira", "marks": [78, 82, 80]},
    {"id": 3, "name": "anitha", "marks": [92, 95, 94]},
    {"id": 4, "name": "nivya", "marks": [70, 75, 72]},
    {"id": 5, "name": "alora", "marks": [88, 85, 90]}
]
for student in students:
    avg = sum(student["marks"]) / len(student["marks"])
    print(student["name"], "Average Marks:", avg)
```

find highest marks

```python
students = [
    {"id": 1, "name": "sreerekha", "marks": [85, 90, 88]},
    {"id": 2, "name": "athira", "marks": [78, 82, 80]},
    {"id": 3, "name": "anitha", "marks": [92, 95, 94]},
    {"id": 4, "name": "nivya", "marks": [70, 75, 72]},
    {"id": 5, "name": "alora", "marks": [88, 85, 90]}
]

top_student = None
highest_avg = 0

for student in students:
    avg = sum(student["marks"]) / len(student["marks"])

    if avg > highest_avg:
        highest_avg = avg
        top_student = student["name"]

print("Top Student:", top_student)
print("Highest Average Marks:", highest_avg)
```

print students with marks >75

```python
students = [
    {"id": 1, "name": "sreerekha", "marks": [85, 90, 88]},
    {"id": 2, "name": "athira", "marks": [78, 82, 80]},
    {"id": 3, "name": "antha", "marks": [92, 95, 94]},
    {"id": 4, "name": "nivya", "marks": [70, 75, 72]},
    {"id": 5, "name": "alora", "marks": [88, 85, 90]}
]

for student in students:
    avg = sum(student["marks"]) / len(student["marks"])

    if avg > 75:
        print(student["name"], "Average:", avg)
```

Task 04 module pratice

use math module


*   squareroot
*   power

use random module


*   random number
*   random choice





```python
import math

# Square root
num = 25
sqrt_value = math.sqrt(num)
print("Square root of", num, "is:", sqrt_value)

# Power
base = 2
exponent = 3
power_value = math.pow(base, exponent)
print("Power:", base, "^", exponent, "=", power_value)
```

```python
import random

# Random number between 1 and 10
rand_num = random.randint(1, 10)
print("Random number:", rand_num)

# Random choice from a list
fruits = ["apple", "banana", "mango", "orange"]
choice = random.choice(fruits)
print("Random choice:", choice)
```

second highest number in list

```python
numbers = [10, 20, 4, 45, 99, 99]
unique_numbers = list(set(numbers))
unique_numbers.sort()

second_highest = unique_numbers[-2]
print("Second highest number:", second_highest)
```

count frequency of characters

```python
text = "hello world"

freq = {}

for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)
```

remove duplicates

```python
numbers = [1, 2, 2, 3, 4, 4, 5]

unique = list(set(numbers))
print(unique)
```
