 difference between if and loop ?

 An if statement checks a condition once. If the condition is true, it runs the code inside and then moves on. If it's false, it skips it.
 A loop (like while or for) checks a condition and runs the code inside repeatedly as long as that condition remains true. It only stops when the condition becomes false.

function to check prime

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
            return True
print(is_prime(7))
```

function to find factorial

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
print(factorial(5))
```

```python
text = input("enter a word :")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

count vowels

```python
text = "Hello World"
vowels = "aeiou"
count = 0

for char in text.lower():
    if char in vowels:
        count += 1

print("Number of vowels:", count)
```
